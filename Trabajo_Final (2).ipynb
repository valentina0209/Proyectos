```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_csv("recdemand.csv")
```

```python
df.info()
```

###**El diccionario de la base de datos implementado para el modelo es:**

trips : número de viajes en bote realizados en el lago Somerville

quality : percepción subjetiva de la calidad de 1 a 5

ski factor: pregunta acerca de si el individuo hizo o no water-ski en el lago Somerville

income: ingreso anual en miles de dólares 

userfee factor: pago de suscripción anual al lago 

costC: costo de visitar Lago Conroe (alternativa al lago Somerville) 

costS: costo de visitar Lago Somerville 

costH: costo de visitar Lago Houston (alternativa al lago Somerville) 

```python
df = df[["trips", "quality","ski","income","userfee","costC", "costS", "costH"]]
```

```python
df
```

```python
df = df.replace({"ski" :{"yes" :1, "no":0 }})
df = df.replace({"userfee" :{"yes" :1, "no":0 }})
```

```python
df
```

#**Modelamiento Regresión Lineal**

```python
y = df["trips"]
x = df[["quality","ski","income","userfee","costC", "costS", "costH"]]
x = sm.add_constant(x)
model = sm.OLS(y,x)
reg = model.fit()
reg.summary()
```

###**Podemos observar que las variables significativas con un P-Valor menor a 0.05 y que en el intervalo de confianza no contienen al cero, son:**
1. quality: que es la percepción subjetiva de la calidad de 1 a 5, con un P-Valor de 0.000
2. ski: que es la pregunta acerca de si hizo o no water-ski en el lago Somerville, con un P-Valor de 0.031  
3. userfee: que es el pago de subcripción anual al lago, con un P-Valor de 0.000
4. costS: Costo de visitar Lago Somerville, con un P-Valor de 0.000 
5. costH: Costo de visitar Lago Houston (alternativa), con un P-Valor de 0.002 

Cabe resaltar que un P-Valor menor al 5% me indica que se rechaza la Hipótesis Nula, en estos casos.

###**Las variables no significativas, es decir que tienen un P-Valor superior al 0.05 son:**

1. Income: que se refiere al ingreso anual en miles de dólares, con un P-Valor de 0.083
3. costC: que se refiere a la alternativa de viajar al lago Somerville, con un P-Valor de 0.066

Cabe resaltar que un P-Valor mayor al 5% me indica que se acepta la Hipótesis Nula, en estos casos.
